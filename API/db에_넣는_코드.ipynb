```python
import mysql.connector
from sqlalchemy import create_engine
```

```python
# 데이터베이스 연결 설정
config = {
    'user': '사용자명넣기',
    'password': '사용자비번넣기',
    'host': '호스트넣기',
    'port': 3306,
    'database': '데이터베이스명'
}

# MySQL 서버에 연결
cnx = mysql.connector.connect(**config)

# SQLAlchemy 엔진 생성
engine = create_engine('mysql+mysqlconnector://사용자명:비번@호스트:3306/f데이터베이스명')
```

```python
# df 데이터프레임을 MySQL에 테이블로 쓰기
df.to_sql(name='설정하고싶은 테이블명', con=engine, if_exists='append', index=False)
# 연결 닫기
cnx.close()
```
